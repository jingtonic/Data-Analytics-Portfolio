### Part 1

Part 1 of the assignment connects to Marvel's API to extract the desired information. The following cells explain what the code does and return the output of the requests.

<div class="alert alert-warning">
    First thing we do is defining our <b>publicKey</b> and <b>privateKey</b> provided by Marvel after we signed in.
</div>

```python
publicKey = '1192095eac208b6580dc631ea767689a'
privateKey = '44fcd1f8c7d2fa4447c7dc2646751f84128efc86'
```

<div class="alert alert-warning">
    Before being able to do a <i>get.request</i> to the API we need to obtain a hash as identification. This hash is formed by the addition of a <b>ts</b> (Timestamp) that we can simply define as <i>= 1</i>, our <b>privateKey</b> and our <b>publicKey</b>. We need to save them in byte form. Then using the function <b>hexdigest()</b>, from hashlib dictionary, we obtain our hash.
</div>

```python
import hashlib

ts = '1'
m = hashlib.md5()
m.update(bytes(ts, 'utf-8'))
m.update(bytes(privateKey, 'utf-8'))
m.update(bytes(publicKey, 'utf-8'))
hash_16 = m.hexdigest()
```

<div class="alert alert-warning">
    Now we are ready to do the <i>get.request</i>. The URL for obatining charachters information is <b>http://gateway.marvel.com/v1/public/characters</b>. The <b>params</b> are our <i>publicKey</i>, the <i>hash</i> created in the above cell, the same <i>timestamp</i> as before and in this case we set a <i>limit = 30</i> in order to obtain just 30 charachters.
</div>

```python
import requests
marvel_url = 'http://gateway.marvel.com'
params = {"apikey": publicKey, "hash": hash_16, 'limit': 30, 'ts': '1'}
headers = {'Accept': '*/*'}
url_charachters = '/v1/public/characters'
characters30 = requests.get(marvel_url + url_charachters, params = params, headers = headers).json()
characters30
```

<div class="alert alert-warning">
    To obtain a list with 30 charachter names we do a <b>for</b> loop that gets the name from the dictionary returned in the request above.
</div>

```python
characters_name = []
for personajes in characters30['data']['results']:
    characters_name += [personajes['name']]
characters_name
```

<div class="alert alert-warning">
    To obtain the <i>id</i> of the charachters we do a <b>for</b> loop that gets the <i>id</i> in a similar way we obtained the names. In this case the <i>id</i> is stored in a dictionary where the keys are the names of the characters.
</div>

```python
characters_id = {}
for i in range(len(characters30['data']['results'])):
    characters_id[characters_name[i]] = str(characters30['data']['results'][i]['id'])
characters_id
```

<div class="alert alert-warning">
    To obtain the <i>total_available_events</i> of the charachters we do a <b>for</b> loop that gets the <i>total_available_events</i> in a similar way we obtained the names. In this case the <i>total_available_events</i> is stored in a dictionary where the keys are the names of the characters.
</div>

```python
total_available_events = {}
for i in range(len(characters30['data']['results'])):
    total_available_events[characters_name[i]] = characters30['data']['results'][i]['events']['available']
total_available_events
```

<div class="alert alert-warning">
    To obtain the <i>total_available_series</i> of the charachters we do a <b>for</b> loop that gets the <i>total_available_series</i> in a similar way we obtained the names. In this case the <i>total_available_series</i> is stored in a dictionary where the keys are the names of the characters.
</div>

```python
total_available_series = {}
for i in range(len(characters30['data']['results'])):
    total_available_series[characters_name[i]] = characters30['data']['results'][i]['series']['available']
total_available_series
```

<div class="alert alert-warning">
    To obtain the <i>total_available_comics</i> of the charachters we do a <b>for</b> loop that gets the <i>total_available_comics</i> in a similar way we obtained the names. In this case the <i>total_available_comics</i> is stored in a dictionary where the keys are the names of the characters.
</div>

```python
total_available_comics = {}
for i in range(len(characters30['data']['results'])):
    total_available_comics[characters_name[i]] = characters30['data']['results'][i]['comics']['available']
total_available_comics
```

<div class="alert alert-warning">
    To obtain the <i>most expensive comic</i> of the charachters we have to change the <i>URL</i> of our <i>requests.get</i>. In this case, the <i>URL</i> changes for every character, therefore we need to do a <b>for</b> loop that takes the <i>id</i> of each charachter from the above created dictionary of ids and pastes it in the <i>URL</i>. Now we do two <b>for</b> loops. The first one returns a dictionary with the different prices for each comic of the character. The second one takes each of these prices. An <b>if</b> checks if the price is the highest one obtained so far, is so, it stores it in a variable <b>price</b>. Once the loops have finished, the highest price is stored in a dictionary, the keys of this dictionary are the names of the characters and the values are the highest prices.
    One important thing to note is that the maximum comics that the requests return is 100. Therefore if there is a character with more than 100 comics we will need to do several requests for this character. We do this with a <b>while</b> loop. The loop takes the total number of existing comics, and each time a loop is completed it substracts 100 from this number. When the value is 0 or lower, the loop stops.
</div>

```python
marvel_url = 'http://gateway.marvel.com'
headers = {'Accept': '*/*'}

comics_max_prices = {}
for keys in characters_id:
    remainig_comics = total_available_comics[keys]
    offset = 0
    price = 0
    while remainig_comics > 0:
        comic_url = marvel_url + '/v1/public/characters/' + characters_id[keys] + '/comics'
        params = {"apikey": publicKey, "hash": hash_16, 'ts': '1', 'limit': 100, 'offset': offset}
        characters_comics = requests.get(comic_url, params = params, headers = headers).json()
        for comics in characters_comics['data']['results']:
            for precios in comics['prices']:
                if precios['price'] > price:
                    price = precios['price']
        offset += 100
        remainig_comics += -100
    comics_max_prices[keys] = price
comics_max_prices
```

<div class="alert alert-warning">
    Now we create a dataFrame with the information obtained in the previous cells using the <i>pandas</i> dictionary . 0 and 0.0 values are substituted with <b>None</b> values.
</div>

```python
import pandas as pd

all_dictionaries = [characters_id, total_available_events, total_available_series,
        total_available_comics, comics_max_prices]

data = characters_name
df = pd.DataFrame(data, columns = ['Character Name'])
df['Character ID'] = df['Character Name'].map(characters_id)
df['Total Available Events'] = df['Character Name'].map(total_available_events)
df['Total Available Series'] = df['Character Name'].map(total_available_series)
df['Total Available Comics'] = df['Character Name'].map(total_available_comics)
df['Price of the Most Expensive Comic'] = df['Character Name'].map(comics_max_prices)
df.replace({0: None, 0.0: None}, inplace = True)
df
```

<div class="alert alert-warning">
    Finally we export the dataFrame to a .csv file. We specify that we do not want the first column to be indexes with the <i>index=False</i> argument.
</div>

```python
df.to_csv('data.csv', index=False)
```
